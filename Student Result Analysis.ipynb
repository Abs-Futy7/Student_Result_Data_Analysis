```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("student_scores.csv")
print(df.head())
```

```python
df.describe()
```

```python
df.info()  # Shows the values which are not null
```

```python
df.isnull().sum() # Showa which values are null
```

# Drop unnamed column

```python
df = df.drop("Unnamed: 0", axis = 1)
print(df.head())
```

# Gender distribution

```python
sns.countplot(data = df, x = "Gender")
plt.show()
```

```python
ax = sns.countplot(data = df, x = "Gender")
plt.title("Gender Distribution")
ax.bar_label(ax.containers[0])
plt.show()
```

```python
## from the above chart we analyzed that:
### the number of females is more than the number of males
```

```python
gb = df.groupby("ParentEduc").agg({"MathScore":'mean',
                                  "ReadingScore": 'mean',
                                  "WritingScore": 'mean'})
print(gb)
```

```python
sns.heatmap(gb, annot = True)
plt.title("Relationship between Parents Education & Student score")
plt.show()
```

```python
#from the above chart we have concluded :
## the education of the parents have a good impact on their scores
```

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',
                                  "ReadingScore": 'mean',
                                  "WritingScore": 'mean'})
print(gb1)
```

```python
sns.heatmap(gb1, annot = True)
plt.title("Relationship between Parent Marital Status & Student score")
plt.figure(figsize= (6,6))
plt.show()
```

```python
#from tha above chart we concluded that:
# parents marital status have negligible impact on students score
```

```python
sns.boxplot(data=df[["ReadingScore", "MathScore", "WritingScore"]])
plt.show()
```

```python
# from the graph we can analyze, the students are weaker in math
```

# Distribution of Ethnic Groups

```python
print(df["EthnicGroup"].unique())
```

```python
groupA = df.loc[(df["EthnicGroup"] == "group A")].count()
groupB = df.loc[(df["EthnicGroup"] == "group B")].count()
groupC = df.loc[(df["EthnicGroup"] == "group C")].count()
groupD = df.loc[(df["EthnicGroup"] == "group D")].count()
groupE = df.loc[(df["EthnicGroup"] == "group E")].count()

l = ["group A", "group B", "group C", "group D", "group E"]
mlist = [groupA["EthnicGroup"], groupB["EthnicGroup"], groupC["EthnicGroup"], groupD["EthnicGroup"], groupE["EthnicGroup"]]
plt.pie(mlist, labels = l, autopct = "%1.2f%%")
plt.title("Ethnic Group Dist.")
plt.show()
```
